```python
def detect_outliers_iqr(df, numeric_columns):
    outlier_indices = []

    for column in numeric_columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1

        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)].index
        outlier_indices.extend(outliers)

    return df.loc[list(set(outlier_indices))]

numeric_columns = [
    "Math_S11", "Critical Reading_S11", "Citizen Competencies_S11", "Biology_S11", "English_S11",
    "Quantitative Reasoning_PRO", "Critical Reading_PRO", "Citizen Competencies_PRO", "English_PRO",
    "Written Communication_PRO", "Formulation of Engineering Projects_PRO", "PERCENTILE",
    "2ND_DECILE", "QUARTILE", "Socioeconomic Level",
    "Socioeconomic Level of The Institution of Higher Education"
]

outliers = detect_outliers_iqr(df, numeric_columns)
print("Outliers:")
print(outliers)
```

## Feature Selection

```python
def get_top_academic_features(df, academic_features, target='G_SC', top_n=10):
    correlations = df[academic_features + [target]].corr()[target].drop(target)
    top_features = correlations.abs().sort_values(ascending=False).head(top_n).index.tolist()
    return top_features

academic_features = [
    'Quantitative Reasoning_PRO', 
    'Critical Reading_PRO', 'Citizen Competencies_PRO', 'English_PRO',
    'Written Communication_PRO', 'Formulation of Engineering Projects_PRO','UNIVERSITY_ENCODED','ACADEMIC_PROGRAM_ENCODED'
]
top_academic = get_top_academic_features(df, academic_features)
print("Top Academic Features:", top_academic)
```

```python
from sklearn.feature_selection import mutual_info_regression
import pandas as pd

def get_top_features_mi(df, features, target='G_SC', top_n=10):
    X = df[features]
    y = df[target]
    X = X.dropna()
    y = y.loc[X.index]
    mi = mutual_info_regression(X, y, discrete_features='auto') 
    mi_series = pd.Series(mi, index=features).sort_values(ascending=False)
    
    return mi_series.head(top_n).index.tolist()

socioeconomic_features = [
    'GENDER', 'Asset Index', 'Revenue','STRATUM_NUMERICAL','OCC_FATHER_ENCODED','OCC_MOTHER_ENCODED','EDU_FATHER_ENCODED','EDU_MOTHER_ENCODED', 'PEOPLE_HOUSE_0', 'PEOPLE_HOUSE_One',
    'PEOPLE_HOUSE_Two', 'PEOPLE_HOUSE_Three', 'PEOPLE_HOUSE_Four', 'PEOPLE_HOUSE_Five', 'PEOPLE_HOUSE_Six',
    'PEOPLE_HOUSE_Seven', 'PEOPLE_HOUSE_Eight', 'PEOPLE_HOUSE_Nine', 'PEOPLE_HOUSE_Ten', 'PEOPLE_HOUSE_Twelve or more','SCHOOL_NAT_PUBLIC',
    'SCHOOL_TYPE_ACADEMIC','SCHOOL_TYPE_Not apply','SCHOOL_TYPE_TECHNICAL','SCHOOL_TYPE_TECHNICAL/ACADEMIC','JOB_0',
    'JOB_No','JOB_Yes, 20 hours or more per week','JOB_Yes, less than 20 hours per week'
]

top_socioeconomic_features = get_top_features_mi(df, socioeconomic_features, target='G_SC', top_n=5)
print("Top Socioeconomic Features:", top_socioeconomic_features)
```

## Model Building

```python
import numpy as np

class MultipleLinearRegression:
    def __init__(self):
        self.coefficients = None

    def fit(self, X, y):
        X_b = np.c_[np.ones((X.shape[0], 1)), X]
        self.coefficients = np.linalg.inv(X_b.T @ X_b) @ X_b.T @ y

    def predict(self, X):
        X_b = np.c_[np.ones((X.shape[0], 1)), X]
        return X_b @ self.coefficients
```

```python
import seaborn as sns

class ModelEvaluator:
    def __init__(self):
        pass

    def mean_squared_error(self, y_true, y_pred):
        return np.mean((y_true - y_pred) ** 2)

    def r2_score(self, y_true, y_pred):
        ss_res = np.sum((y_true - y_pred) ** 2)
        ss_tot = np.sum((y_true - np.mean(y_true)) ** 2)
        return 1 - (ss_res / ss_tot)

    def plot_predictions(self, y_true, y_pred, title='Actual vs Predicted'):
        plt.figure(figsize=(8, 6))
        plt.scatter(y_true, y_pred, color='blue', alpha=0.5, label='Predictions')
        plt.plot([y_true.min(), y_true.max()], [y_true.min(), y_true.max()], color='red', linestyle='--', label='Ideal (y = x)')
        plt.xlabel('Actual G_SC')
        plt.ylabel('Predicted G_SC')
        plt.title(title)
        plt.grid(True)
        plt.legend()
        plt.tight_layout()
        plt.show()

    def plot_residuals(self, residuals, title='Residuals Distribution'):
        plt.figure(figsize=(8, 6))
        # Plotting histogram and kernel density estimate (KDE)
        sns.histplot(residuals, kde=True, color='purple', bins=30, stat='density')
        plt.xlabel('Residuals')
        plt.ylabel('Density')
        plt.title(title)
        plt.grid(True)
        plt.tight_layout()
        plt.show()
        
```

```python
class EnhancedPerformancePredictor:
    def __init__(self, df):
        self.df = df.copy()
        self.model = MultipleLinearRegression()
        self.evaluator = ModelEvaluator()
        self.academic_features = [
            top_academic[0],
            top_academic[1],
            top_academic[2],
            top_academic[3],
            top_academic[4],
            top_academic[5]
        ]
        self.socioeconomic_features = [
            top_socioeconomic[0],
            top_socioeconomic[1],
            top_socioeconomic[2],
            top_socioeconomic[3],
            top_socioeconomic[4],
            top_socioeconomic[5]
        ]
    
    def preprocess(self):
        self.df = self.df.dropna()
        # No missing values, minimal preprocessing needed
        pass

    def train_and_evaluate(self, features):
        X = self.df[features].values
        y = self.df['G_SC'].values

        # Train-test split manually (80% train, 20% test)
        n = len(X)
        split_idx = int(0.8 * n)
        X_train, X_test = X[:split_idx], X[split_idx:]
        y_train, y_test = y[:split_idx], y[split_idx:]

        # Train the model
        self.model.fit(X_train, y_train)

        # Predictions
        y_train_pred = self.model.predict(X_train)
        y_test_pred = self.model.predict(X_test)

        # Training metrics
        train_mse = self.evaluator.mean_squared_error(y_train, y_train_pred)
        train_r2 = self.evaluator.r2_score(y_train, y_train_pred)

        # Test metrics
        test_mse = self.evaluator.mean_squared_error(y_test, y_test_pred)
        test_r2 = self.evaluator.r2_score(y_test, y_test_pred)

        print(f"Features used: {features}")
        print("\n--- Training Performance ---")
        print(f"Train MSE: {train_mse:.4f}")
        print(f"Train R²: {train_r2:.4f}")

        print("\n--- Test Performance ---")
        print(f"Test MSE: {test_mse:.4f}")
        print(f"Test R²: {test_r2:.4f}")

        # Plotting for test set
        self.evaluator.plot_predictions(y_test, y_test_pred, title=f'Actual vs Predicted (Test Set) - {features}')
        residuals = y_test - y_test_pred
        self.evaluator.plot_residuals(residuals, title=f'Residuals (Test Set) - {features}')

        return train_r2, test_r2  # Return both if needed

```

```python
predictor = EnhancedPerformancePredictor(df)

predictor.preprocess()

# Drop any rows with NaN values before evaluation
df_clean = df.dropna()

print("------- Academic Factors -------")
train_r2_academic, test_r2_academic = predictor.train_and_evaluate(predictor.academic_features)

print("\n------- Socioeconomic Factors -------")
train_r2_socio, test_r2_socio = predictor.train_and_evaluate(predictor.socioeconomic_features)

# Compare using test R² scores
if test_r2_academic > test_r2_socio:
    print("\n Academic factors explain G_SC better.")
    print(f"Test R² Academic: {test_r2_academic:.4f} vs Test R² Socioeconomic: {test_r2_socio:.4f}")
else:
    print("\n Socioeconomic factors explain G_SC better.")
    print(f"Test R² Socioeconomic: {test_r2_socio:.4f} vs Test R² Academic: {test_r2_academic:.4f}")
```
